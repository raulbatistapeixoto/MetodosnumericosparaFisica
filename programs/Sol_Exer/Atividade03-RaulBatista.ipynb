1 - A força magnética sobre uma partícula de massa m e carga q é: $\vec{F_{B}} = q(\vec{v} \times \vec{B})$

* 1.a - Escreva explicitamente as equações do movimento para $x$,$y$ e $z$ e para $v_x$, $v_y$ e $v_z$.

### Solução:

Temos a equação de movimento: 

$$\frac{d\vec{v}}{dt} =  \frac{q}{m}(\vec{v} \times \vec{B})$$

Nos fornece as equaçãoes: 

$$\frac{dv_{x}}{dt} =  \frac{q}{m}v_{y}B$$

$$\frac{dv_{y}}{dt} =  -\frac{q}{m}v_{x}B$$

$$\frac{dv_{z}}{dt} =  0$$

Observe que teremos:

$$\frac{d^{2}v_{x}}{dt^{2}} = -\left (\frac{qB}{m}\right)^{2}v_{x}$$

A solução é dada por:

$$v_x(t) = -v\sin(\omega t + \phi_{0})$$

onde $\omega = \frac{qB}{m}$ e $v^2 = v_{x}^2 + v_{y}^2 + v_{z}^2$.

Podemos agora calcular a componente y da velocidade:

$$v_y(t) = -v\cos(\omega t + \phi_{0})$$

Integrando as expresões para as velocidades teremos:

$$x(t) = x_{0} + \frac{v}{\omega}\cos(\omega t + \phi_{0})$$
$$y(t) = y_{0} - \frac{v}{\omega}\sin(\omega t + \phi_{0})$$
$$z(t) = z_{0} + v_{z0}t $$

Deste forma obtemos a lei horária de movimentos para as coordenadas x,y e z.

* 1.b - Determine a versão discretizada destas equações.

### Solução:
Assumindo a derivada progressiva temos: 

$$\frac{d x}{dt} \approx \frac{x(t+\Delta t) - x(t)}{\Delta t}$$

$$\frac{d v}{dt} \approx \frac{v(t+\Delta t) - v(t)}{\Delta t}$$

Sendo assim a forma discretizada das equações diferencias se torna:

$$x_{i+1} = x_i + v_{x,i} \Delta t$$

$$v_{x,i+1} = v_{x,i} + a_{x,i} \Delta t$$

$$v_{x,i+1} = v_{x,i} + \frac{q}{m}Bv_{y,i} \Delta t$$

$$y_{i+1} = y_i + v_{y,i} \Delta t$$

$$v_{y,i+1} = v_{y,i} + a_{y,i} \Delta t$$

$$v_{y,i+1} = v_{y,i} - \frac{q}{m}Bv_{x,i} \Delta t$$

$$v_y(t) = -v\cos(\omega t + \phi_{0}) =  v_{y0}\cos(\omega t) - v_{x0}\sin(\omega t)$$

onde $\omega = \frac{qB}{m}$. 

Para as coordenadas temos:

$$y(t) = y_{0} - \frac{v}{\omega}\sin(\omega t + \phi_{0}) = y_{0} + \frac{v_{y0}}{\omega}\cos(\omega t) - \frac{v_{x0}}{\omega}\sin(\omega t)$$

```python
# assigning input variables

dt = 0.01
nt = 5000
q = 1
m = 1
x0 = 0
y0 = 5
z0 = 2
vx0 = 5
vy0 = 0
vz0 = 0
B = 1
omega = ((B*q)/m)
        
```

```python
# defining a function that calculates the whole 
# time iteration of the diffencial equation 

def magCharg(x,y,z,vx,vy,vz,omega):

    
    # array creation for all quantities
    time = np.zeros(nt)
    x = np.zeros(nt)    # array of x coordinates    
    y = np.zeros(nt)    # array of y coordinates 
    z = np.zeros(nt)    # array of z coordinates 
    vx = np.zeros(nt)   # array of vx 
    vy = np.zeros(nt)   # array of vy 
    vz = np.zeros(nt)   # array of vz 

    # defining initial conditions
    x[0] = x0
    y[0] = y0
    z[0] = z0
    vx[0] = vx0
    vy[0] = vy0
    vz[0] = vz0

    # iterating loop starting at second array position 
    for i in range(1,nt):
        time[i] = time[i-1] + dt
        x[i] = x[i-1] + vx[i-1]*dt
        y[i] = y[i-1] + vy[i-1]*dt
        z[i] = z[i-1] + vz[i-1]*dt
        vx[i] = vx[i-1] + vy[i-1]*omega*dt
        vy[i] = vy[i-1] - vx[i-1]*omega*dt
        vz[i] = vz[i-1]
    
    return x,y,z,vx,vy,vz,time
```

Note que este modelo de atrito é mais simples que o modelo devido a resistência do ar. Descreva explicitamente as equações diferencias discretizadas para este problema

## Solução:

Note que considerando $\vec{B} = (0,0,B) $ teremos as equações:

$$\frac{dv_{x}}{dt} =  \frac{q}{m}v_{y}B - \frac{\gamma}{m} v_x$$

$$\frac{dv_{y}}{dt} =  -\frac{q}{m}v_{x}B - \frac{\gamma}{m} v_y$$

$$\frac{dv_{z}}{dt} =  -\frac{\gamma}{m} v_z $$

A forma discreta destas equações nos dá:

$$x_{i+1} = x_i + v_{x,i} \Delta t$$

$$v_{x,i+1} = v_{x,i} + a_{x,i} \Delta t$$

$$v_{x,i+1} = v_{x,i}(1 - \frac{\gamma}{m}  \Delta t) + \frac{q}{m}Bv_{y,i} \Delta t$$

$$y_{i+1} = y_i + v_{y,i} \Delta t$$

$$v_{y,i+1} = v_{y,i} + a_{y,i} \Delta t$$

$$v_{y,i+1} = v_{y,i}(1 - \frac{\gamma}{m} \Delta t) - \frac{q}{m}Bv_{x,i} \Delta t$$

$$z_{i+1} = z_i + v_{z,i} \Delta t$$

$$v_{z,i+1} = v_{z,i} + a_{z,i} \Delta t$$

$$v_{z,i+1} = v_{z,i}(1 - \frac{\gamma}{m}  \Delta t) $$

A forma exata para as velocidades é dada por:

$$v_x = e^{-\frac{\gamma}{m}t}\left(v_{0x}\cos(\omega t) + \frac{v_{0y}}{\omega}\sin(\omega t) \right)$$

$$v_y = e^{-\frac{\gamma}{m}t}\left(v_{0y}\cos(\omega t) - \frac{v_{0x}}{\omega}\sin(\omega t) \right)$$

$$v_z = v_{0z}e^{-\frac{\gamma}{m}}$$

Para as coordenadas teremos:

$$x = x_0 + \frac{v_{0x}}{\left( \frac{\gamma^2}{m^2} + \omega^2 \right)}e^{-\frac{\gamma}{m}}\left( \omega \sin(\omega t ) - \frac{\gamma}{m}\cos(\omega t) + \frac{\gamma}{m} \right) - \frac{v_{0y}}{\omega \left( \frac{\gamma^{2}}{m^2} + \omega^2 \right)}e^{-\frac{\gamma}{m}}\left( \omega \sin(\omega t ) + \frac{\gamma}{m}\cos(\omega t) + \omega \right)$$

$$ = y_0 + \frac{v_{0y}}{\left( \frac{\gamma^2}{m^2} + \omega^2 \right)}e^{-\frac{\gamma}{m}}\left( \omega \sin(\omega t ) - \frac{\gamma}{m}\cos(\omega t) + \frac{\gamma}{m} \right) + \frac{v_{0x}}{\omega \left( \frac{\gamma^{2}}{m^2} + \omega^2 \right)}e^{-\frac{\gamma}{m}}\left( \omega \sin(\omega t ) + \frac{\gamma}{m}\cos(\omega t) + \omega \right)$$

$$z = z_0 - \frac{v_{0z}}{\left(\frac{\gamma}{m}\right)}\left(e^{-\frac{\gamma}{m}} - 1  \right)$$

```python
# loading libraries 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
# assigning input variables

dt = 0.01
nt = 5000
q = 1
m = 1
gamma = 0.1
x0 = 1
y0 = 0
z0 = 0
vx0 = 0
vy0 = 1
vz0 = 0
Bz = 1
By = 0
Bx = 0
omega = ((Bz*q)/m)
beta = (gamma/m)
alpha = (gamma**2 + omega**2)
        
```

```python
# defining a function that calculates the whole 
# time iteration of the diffencial equation 

def magCharg(x,y,z,vx,vy,vz,omega):

    
    # array creation for all quantities
    time = np.zeros(nt)
    x = np.zeros(nt)    # array of x coordinates    
    y = np.zeros(nt)    # array of y coordinates 
    z = np.zeros(nt)    # array of z coordinates 
    vx = np.zeros(nt)   # array of vx 
    vy = np.zeros(nt)   # array of vy 
    vz = np.zeros(nt)   # array of vz 

    # defining initial conditions
    x[0] = x0
    y[0] = y0
    z[0] = z0
    vx[0] = vx0
    vy[0] = vy0
    vz[0] = vz0

    # iterating loop starting at second array position 
    for i in range(1,nt):
        time[i] = time[i-1] + dt
        x[i] = x[i-1] + vx[i-1]*dt
        y[i] = y[i-1] + vy[i-1]*dt
        z[i] = z[i-1] + vz[i-1]*time[i]
        vx[i] = vx[i-1] - vx[i-1]*(gamma/m)*dt + vy[i-1]*(omega)*dt
        vy[i] = vy[i-1] - vy[i-1]*(gamma/m)*dt - vx[i-1]*(omega)*dt
        vz[i] = vz[i-1] - vz[i-1]*(gamma/m)*dt
    
        
    return x,y,z,vx,vy,vz,time
```

```python
# Cálculo da solução exata:

def funcMag(x,y,z,vx,vy,vz,t,omega):
    
    x = x0 + (vx0/(alpha))*omega*np.sin(omega*t)*np.exp(-beta*t) - (vx0/alpha)*beta*np.cos(omega*t)*np.exp(-beta*t) - (vy0/alpha)*np.cos(omega*t)*np.exp(-beta*t) - (vy0/(alpha*omega))*beta*np.sin(omega*t)*np.exp(-beta*t) + (vx0/(alpha))*beta + (vy0/(alpha))         
    y = y0 + (vy0/(alpha))*omega*np.sin(omega*t)*np.exp(-beta*t) - (vy0/alpha)*beta*np.cos(omega*t)*np.exp(-beta*t) + (vx0/alpha)*np.cos(omega*t)*np.exp(-beta*t) + (vx0/(alpha*omega))*beta*np.sin(omega*t)*np.exp(-beta*t)  + (vy0/(alpha))*beta - (vx0/(alpha))
    z = z0 + (vz0/beta) - (vz0/beta)*np.exp(-beta*t)
    
    return x,y,z
```

```python
# it only calculates the total time 
Tmax = (nt-1)*dt 

# cria lista dos t_i com Nt elementos todos inicialmente zerados
time = np.linspace(0, Tmax, nt)

# calculating the solution 
x,y,z,vx,vy,vz,time = magCharg(x0,y0,z0,vx0,vy0,vz0,omega)
xt,yt,zt = funcMag(x0,y0,z0,vx0,vy0,vz0,time,omega)
```

```python
# Gráficos da solução numérica:

plt.plot(x,y,'b-')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend(['Sol. Numérica x,y'])
plt.show()

plt.plot(time,z,'m-')
plt.xlabel('t (s)')
plt.ylabel('z (m)')
plt.legend(['Sol. Numérica t,z'])
plt.show()
```

```python
fig = plt.figure(figsize=(15, 7.5))

# VISAO OBLIQUA
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot(x, y, z, 'y')
ax.view_init(elev=40., azim=30) 

# VISAO SUPERIOR
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot(x, y, z, 'm')
ax.view_init(elev=90., azim=90) 

plt.show()
```

```python
# Gráficos solução Exata:

plt.plot(xt,yt,'y-')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend(['Sol. Exata x,y'])
plt.show()

plt.plot(time,zt,'m-')
plt.xlabel('t (s)')
plt.ylabel('z (m)')
plt.legend(['Sol. Numérica t,z'])
plt.show()
```

```python
# Função Exata:

fig = plt.figure(figsize=(15, 7.5))

# VISAO OBLIQUA
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot(xt, yt, zt, 'r')
ax.view_init(elev=40., azim=10) 

# VISAO SUPERIOR
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot(xt, yt, zt, 'b')
ax.view_init(elev=90., azim=90) 

plt.show()
```
